# Section 2 — Data Preprocessing & Cleaning

### 2.1 Cek Missing Values

```python
print("=== MISSING VALUES - ORDERS ===")
print(orders.isnull().sum())
print()
print("=== MISSING VALUES - ORDER ITEMS ===")
print(order_items.isnull().sum())
print()
print("=== MISSING VALUES - PRODUCTS ===")
print(products.isnull().sum())
```

### Jawaban 2.2 — Analisis Missing Values

**Kolom yang memiliki missing value:**
- `orders`: `order_approved_at` (160), `order_delivered_carrier_date` (1.783), `order_delivered_customer_date` (2.965)
- `products`: `product_category_name` (610) dan beberapa kolom dimensi (2)

---

**Mengapa kolom tanggal delivery di orders bisa kosong?**

Karena tidak semua pesanan sudah sampai ke pelanggan. Pesanan dengan status `canceled`, `processing`, atau `shipped` belum tentu punya tanggal delivery. Jadi missing value di kolom ini **wajar dan expected**, bukan data error.

---

**Keputusan penanganan missing value:**
- `product_category_name` → isi dengan `'Unknown'` karena tetap perlu dianalisis per kategori
- Kolom tanggal delivery di `orders` → biarkan dulu, akan di-filter saat dibutuhkan di EDA
- Kolom dimensi produk (2 baris) → biarkan, tidak dipakai di analisis utama


### 2.3 Tangani Missing Values

```python
# Isi kategori produk yang kosong dengan 'Unknown'
products['product_category_name'] = products['product_category_name'].fillna('Unknown')

print("Missing values products setelah dibersihkan:")
print(products.isnull().sum())
```

### 2.4 Cek Duplikat

```python
print(f"Duplikat di orders      : {orders.duplicated().sum()}")
print(f"Duplikat di order_items : {order_items.duplicated().sum()}")
print(f"Duplikat di products    : {products.duplicated().sum()}")
```

### 2.5 Konversi Tipe Data — DateTime

```python
# Konversi kolom tanggal di orders
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
for col in date_cols:
    orders[col] = pd.to_datetime(orders[col])

# Konversi di order_items
order_items['shipping_limit_date'] = pd.to_datetime(order_items['shipping_limit_date'])

print("Tipe data orders setelah konversi:")
print(orders.dtypes)
```

### 2.6 Merge / Join Tabel

```python
# Step 1: merge orders + order_items
df = pd.merge(orders, order_items, on='order_id', how='left')
print(f"Shape setelah merge pertama: {df.shape}")
```

```python
# Step 2: merge + products (ambil kolom yang dibutuhkan saja)
df = pd.merge(df, products[['product_id', 'product_category_name']], on='product_id', how='left')
print(f"Shape setelah merge kedua (final): {df.shape}")
display(df.head())
```

### Jawaban — Catatan Setelah Merge

Setelah merge, ditemukan **775 baris dengan nilai null di kolom `price`, `freight_value`, dan `product_id`**.

Ini terjadi karena kita menggunakan `how='left'` — semua baris dari `orders` dipertahankan, termasuk orders yang tidak memiliki pasangan di `order_items`. Orders tanpa item berarti tidak ada transaksi yang bisa dianalisis, sehingga baris-baris ini di-drop.

Null yang tersisa di kolom tanggal delivery (`order_delivered_carrier_date`, `order_delivered_customer_date`) adalah **wajar** — pesanan yang belum dikirim atau belum sampai memang tidak punya tanggal tersebut.


```python
# Cek null setelah merge
print("=== NULL SETELAH MERGE ===")
print(df.isnull().sum()[df.isnull().sum() > 0])
print()
print(f"Baris dengan price = null: {df['price'].isnull().sum()}")
print("=> Ini adalah orders yang tidak memiliki item di order_items (left join)")
```

```python
# Drop baris yang tidak punya item (price = null)
# Orders tanpa item tidak bisa dianalisis revenue-nya
before = len(df)
df = df.dropna(subset=['price', 'freight_value', 'product_id'])
after = len(df)

print(f"Baris sebelum drop : {before}")
print(f"Baris setelah drop : {after}")
print(f"Baris di-drop      : {before - after}")
print()
print("=== NULL SETELAH CLEANING FINAL ===")
remaining_nulls = df.isnull().sum()[df.isnull().sum() > 0]
print(remaining_nulls)
print()
print("(Null di kolom tanggal delivery = wajar, order belum sampai)")
```

```python
# Tambah kolom bantu untuk EDA
df['order_month'] = df['order_purchase_timestamp'].dt.to_period('M')
df['month_only'] = df['order_purchase_timestamp'].dt.month
df['year'] = df['order_purchase_timestamp'].dt.year

print("Kolom final:")
print(df.columns.tolist())
```

### Jawaban 2.7 — Refleksi Cleaning & Merge

**Yang dilakukan di section cleaning:**
1. Mengisi missing value `product_category_name` dengan `'Unknown'`
2. Tidak ada duplikat di semua tabel, tidak perlu drop
3. Mengkonversi 5 kolom tanggal di `orders` dan 1 kolom di `order_items` dari `object` ke `datetime`
4. Merge `orders` + `order_items` berdasarkan `order_id` (left join)
5. Merge hasil tersebut + `products` berdasarkan `product_id` (left join)

---

**Shape dataframe final:** sekitar 112.650 baris × 17 kolom. Lebih banyak dari `orders` karena satu order bisa punya lebih dari satu item.

---

**Kenapa perlu merge?**
Tanpa merge, kita tidak bisa tahu produk apa yang ada di setiap pesanan, berapa harganya, dan kategori apa yang paling laku. Ketiga tabel hanya punya potongan informasi — setelah digabung, kita punya gambaran lengkap dari setiap transaksi.

# Section 3 — Exploratory Data Analysis (EDA)

### 3.1 Distribusi Harga Produk

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df['price'].dropna(), bins=50, ax=axes[0], color='steelblue')
axes[0].set_title('Distribusi Harga Produk')
axes[0].set_xlabel('Harga (BRL)')
axes[0].set_ylabel('Jumlah')

sns.boxplot(x=df['price'].dropna(), ax=axes[1], color='lightcoral')
axes[1].set_title('Boxplot Harga Produk')
axes[1].set_xlabel('Harga (BRL)')

plt.tight_layout()
plt.show()
```

### Jawaban Insight 3.1

Distribusi harga sangat **right-skewed** — mayoritas produk dijual di kisaran harga rendah (di bawah BRL 200), namun ada banyak outlier dengan harga sangat tinggi hingga ribuan BRL. Ini umum di platform e-commerce karena ada produk murah (aksesoris, kosmetik) dan produk mahal (elektronik, furnitur) yang dijual bersamaan. Median lebih representatif daripada mean untuk dataset ini.


### 3.2 Top 10 Kategori Berdasarkan Revenue

```python
revenue_by_category = (
    df.groupby('product_category_name')['price']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(revenue_by_category)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=revenue_by_category.values, y=revenue_by_category.index, palette='Blues_r')
plt.title('Top 10 Kategori Produk Berdasarkan Total Revenue')
plt.xlabel('Total Revenue (BRL)')
plt.ylabel('Kategori Produk')
plt.tight_layout()
plt.show()
```

### Jawaban Insight 3.2

Kategori **bed/table/bath (cama_mesa_banho)** dan **sports & leisure (esporte_lazer)** mendominasi revenue. Ini menunjukkan bahwa produk kebutuhan rumah tangga dan gaya hidup adalah segmen terkuat di platform ini. Kategori-kategori ini memiliki volume transaksi tinggi dan harga yang cukup bersaing, sehingga menghasilkan revenue kumulatif yang besar.


### 3.3 Tren Jumlah Pesanan per Bulan

```python
monthly_orders = (
    df.groupby('order_month')['order_id']
    .nunique()
    .reset_index()
    .sort_values('order_month')
)
monthly_orders.columns = ['month', 'total_orders']
monthly_orders['month'] = monthly_orders['month'].astype(str)

plt.figure(figsize=(14, 5))
plt.plot(monthly_orders['month'], monthly_orders['total_orders'],
         marker='o', color='steelblue', linewidth=2, markersize=5)
plt.title('Tren Jumlah Pesanan per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Pesanan')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Jawaban Insight 3.3

Tren pesanan menunjukkan **pertumbuhan signifikan** dari akhir 2016 hingga pertengahan 2018. Puncak terjadi sekitar **November 2017** — kemungkinan besar karena **Black Friday**, yang merupakan event belanja besar di Brasil. Ada penurunan di akhir dataset (pertengahan 2018) yang kemungkinan karena data tidak lengkap untuk bulan-bulan terakhir, bukan karena penurunan penjualan nyata.


### 3.4 Distribusi Status Pesanan

```python
status_counts = df['order_status'].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

status_counts.plot(kind='bar', ax=axes[0], color='steelblue', edgecolor='white')
axes[0].set_title('Distribusi Status Pesanan')
axes[0].set_xlabel('Status')
axes[0].set_ylabel('Jumlah')
axes[0].tick_params(axis='x', rotation=45)

axes[1].pie(status_counts.values, labels=status_counts.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Proporsi Status Pesanan')

plt.tight_layout()
plt.show()
```

### Jawaban Insight 3.4

Lebih dari **96%** pesanan berhasil **delivered** — angka yang sangat baik untuk platform e-commerce. Tingkat cancellation hanya sekitar **0.6%**, menunjukkan sistem fulfillment yang cukup handal. Namun perlu diperhatikan pesanan berstatus `unavailable` (~0.6%) yang artinya produk tidak tersedia saat order dibuat — ini perlu diperbaiki dari sisi manajemen inventori seller.


### 3.5 Rata-rata Harga vs Freight per Kategori

```python
top_categories = revenue_by_category.index.tolist()

price_freight = (
    df[df['product_category_name'].isin(top_categories)]
    .groupby('product_category_name')[['price', 'freight_value']]
    .mean()
)

price_freight.plot(kind='bar', figsize=(14, 6),
                   color=['steelblue', 'lightcoral'], edgecolor='white')
plt.title('Rata-rata Harga vs Freight Value per Kategori (Top 10)')
plt.xlabel('Kategori Produk')
plt.ylabel('Nilai Rata-rata (BRL)')
plt.xticks(rotation=45, ha='right')
plt.legend(['Harga Produk', 'Ongkos Kirim'])
plt.tight_layout()
plt.show()
```

### Jawaban Insight 3.5

Kategori **furniture & decor (moveis_decoracao)** memiliki rasio freight-to-price yang relatif tinggi dibanding kategori lain — wajar karena produk furnitur besar dan berat. Sebaliknya, kategori **beauty & health (beleza_saude)** punya harga rata-rata lebih tinggi dengan ongkir relatif rendah, artinya margin lebih baik. Ini bisa jadi insight bagi seller untuk mengoptimalkan kategori produk yang dijual.


### 3.6 Pivot Table — Revenue per Kategori per Bulan

```python
top_cat_df = df[df['product_category_name'].isin(top_categories[:8])]

pivot = pd.pivot_table(
    top_cat_df,
    values='price',
    index='product_category_name',
    columns='month_only',
    aggfunc='sum',
    fill_value=0
)

display(pivot)
```

```python
plt.figure(figsize=(14, 7))
sns.heatmap(pivot, annot=True, fmt='.0f', cmap='YlOrRd', linewidths=0.5)
plt.title('Revenue per Kategori per Bulan (BRL)')
plt.xlabel('Bulan')
plt.ylabel('Kategori Produk')
plt.tight_layout()
plt.show()
```

### Jawaban Insight 3.6

Dari heatmap terlihat bahwa hampir semua kategori mencapai puncak revenue di bulan **8 (Agustus)** dan **11 (November)**. Bulan November konsisten dengan Black Friday. Bulan Agustus kemungkinan karena musim belanja pertengahan tahun. Kategori **cama_mesa_banho** konsisten tinggi sepanjang tahun, menandakan permintaan yang stabil — bukan hanya musiman.


### 3.7 Crosstab — Status Pesanan vs Bulan

```python
ct = pd.crosstab(df['order_status'], df['month_only'])
ct.columns.name = 'Bulan'
display(ct)
```

```python
main_status = ['delivered', 'canceled', 'shipped']
ct_main = ct.loc[main_status]

ct_main.plot(kind='bar', figsize=(14, 6), edgecolor='white')
plt.title('Distribusi Status Pesanan per Bulan')
plt.xlabel('Status Pesanan')
plt.ylabel('Jumlah')
plt.xticks(rotation=0)
plt.legend(title='Bulan', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

### Jawaban Insight 3.7

Cancellation paling tinggi terjadi di bulan-bulan dengan volume pesanan tinggi (Agustus, November) — ini wajar karena volume total memang lebih besar. Namun jika dilihat persentasenya, cancellation rate tetap konsisten rendah. Pesanan dengan status `shipped` tertinggi di bulan-bulan akhir dataset, kemungkinan karena pesanan tersebut belum sempat ter-update menjadi `delivered`.

## Business Questions


### Business Question 1

**Pertanyaan:** Kategori produk apa yang paling menguntungkan?

**Temuan:**
Kategori **cama_mesa_banho (bed/table/bath)** menghasilkan total revenue tertinggi, diikuti oleh **esporte_lazer (sports & leisure)** dan **moveis_decoracao (furniture & decor)**. Ketiganya secara konsisten unggul di hampir semua bulan sepanjang tahun. Ketiga kategori ini merupakan produk kebutuhan rumah tangga dan gaya hidup yang memiliki permintaan stabil dan volume transaksi tinggi.

**Rekomendasi:**
Platform sebaiknya memberikan prioritas promosi, flash sale, dan program loyalitas seller untuk tiga kategori teratas ini karena kontribusinya terhadap revenue sangat signifikan. Investasi iklan di kategori ini akan memberikan ROI lebih tinggi dibanding kategori ekor.


### Business Question 2

**Pertanyaan:** Kapan puncak penjualan terjadi dan apa penyebabnya?

**Temuan:**
Puncak penjualan terjadi di **November 2017** dengan lonjakan signifikan dibanding bulan-bulan sebelumnya. Secara bulanan, bulan **8 (Agustus)** dan **11 (November)** konsisten menjadi bulan dengan revenue tertinggi di hampir semua kategori. Tren keseluruhan menunjukkan pertumbuhan positif dari Q4 2016 hingga pertengahan 2018.

**Rekomendasi:**
Siapkan stok dan kapasitas logistik lebih awal sebelum November (Black Friday) dan Agustus. Campaign marketing harus dimulai minimal 2–3 minggu sebelum event besar untuk memaksimalkan konversi. Platform juga dapat menciptakan event belanja sendiri di bulan-bulan yang secara historis sepi (Februari, Maret) untuk meratakan distribusi penjualan.


### Business Question 3

**Pertanyaan:** Bagaimana performa pengiriman secara keseluruhan?

**Temuan:**
Tingkat keberhasilan pengiriman (**delivered**) mencapai lebih dari **96%** dari total pesanan — angka yang sangat baik. Cancellation rate hanya sekitar **0.6%**. Namun ada sekitar **0.6% pesanan berstatus unavailable**, yang artinya produk tidak tersedia setelah order dibuat — ini menandakan masalah sinkronisasi stok dari sisi seller.

**Rekomendasi:**
Pertahankan performa pengiriman yang sudah sangat baik. Fokus perbaikan pada pesanan `unavailable` — platform perlu mewajibkan seller untuk selalu memperbarui stok secara real-time dan memberikan penalti jika tingkat unavailability terlalu tinggi, karena ini langsung berdampak pada pengalaman pelanggan.
